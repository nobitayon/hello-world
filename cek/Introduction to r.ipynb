```r
x<- seq(-pi,pi,length=50)
x
```

```r
y<- x
f<- outer(x,y,function(x,y) cos(y) / (1 + x^2))
contour(x,y,f)
contour(x,y,f,nlevels=45,add=T)
```

```r
fa<- (f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
image(x,y,fa)
```

```r
persp(x,y,fa)
persp(x,y,fa,theta=30,main='theta=30')
persp(x,y,fa,theta=30,phi=20,main='theta=30,phi=20')
persp(x,y,fa,theta=30,phi=70,main='theta=30,phi=70')
persp(x,y,fa,theta=30,phi=40,main='theta=30,phi=40')
```
